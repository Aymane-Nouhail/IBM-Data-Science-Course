### The following code will approximate the integral of a function *f* on the i,terval *[a,b]* using the rectangles approach :

```python
def RectangleGauche(f,a,b,n): #n is the number of iterations
    h = (b-a)/n
    ak = a
    res = 0
    for i in range(n):
        res += f(ak)
        ak += h
    return h * res
def f(x):
    return x**2
RectangleGauche(f,1,3,1000)
```
